```python
def clean_sqft(sqft):
    l=sqft.split("-")
    if len(l)==2:
        return float(l[0])+float(l[1])/2
    try:
        return float(l[0])
    except:
        return None

df["total_sqft"]=df["total_sqft"].apply(clean_sqft)
```

```python
df
```

```python
df["total_sqft"]=df["total_sqft"].fillna(round(df["total_sqft"].mean()))
```

```python
df["price_per_sqft"]=df["price"]*100000/df["total_sqft"]
df
```

```python
# removing the outliers present in the price_per_sqrt
q1=df["price_per_sqft"].quantile(0.25)
q3=df["price_per_sqft"].quantile(0.75)
```

```python
df=df.drop(["index","size","price_per_sqft"],axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
encocder = LabelEncoder()
df['encoded_loc'] = encocder.fit_transform(df['location'])
df
```

```python
x = df.drop(['location','price'],axis=1)
y = df.price
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
```

```python
params = {
    'n_estimators':[100,150,200,250,300],
    'max_depth':[3,4,5,6,7]
}
```

```python
Grid = GridSearchCV(estimator=model,param_grid=params,cv=5)
Grid.fit(xtrain,ytrain)
```

```python
print("Best Params",Grid.best_params_)
print("Best score",Grid.best_score_)
```

```python
ypred=Grid.predict(xtest)
ypred
```

```python
print("Training Efficency",Grid.score(xtrain,ytrain))
print("Testing Efficency",Grid.score(xtest,ytest))
```
